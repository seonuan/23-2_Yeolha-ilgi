
### <strong> 박지원의 <<열하일기>> 중 <도강록>에 빈번하게 쓰인 단어는 무엇일까요? </strong>

### 저희는 워드클라우드 기능을 사용해 <도강록>에서의 주요 키워드를 파악했습니다.
###### KoNLPy 출처: 박은정, 조성준, “KoNLPy: 쉽고 간결한 한국어 정보처리 파이썬 패키지”, 제 26회 한글 및 한국어 정보처리 학술대회 논문집, 2014.
###### 출처 : 『한국고전종합DB』, 한국고전번역원

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
import numpy as np
import seaborn as sns
```

```python
with open("도강록.txt", 'r', encoding='utf-8') as f:
    text = f.read()
text
```

### <도강록> 전문을 기반으로 한 워드클라우드 입니다. 
### 불용어를 따로 지정하지 않았으며, KoNLPy가 분리한 명사를 바탕으로 제작했습니다.

```python
# 워드 클라우드 1차 생성 - 원본
okt = Okt()
nouns = okt.nouns(text) # 명사만 사용

words = [n for n in nouns if len(n) > 1]
c = Counter(words)

wc = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, 
    max_font_size=250, background_color='white')
gen = wc.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)
```

```python
okt = Okt()

li_pos_tag=print(okt.pos(text, join=True))
nouns = okt.nouns(text) # 명사만 추출

words = [n for n in nouns if len(n) > 1]
c = Counter(words)

file_name="도강록_명사추출.txt"
with open(file_name, 'w+') as file:
    file.write(','.join(c))
    
c
```

### 위의 결과를 보면 '모두', '마치' 등 의미 없는 단어가 포함되기도 합니다.
### 이번에는 불용어를 총 2회 제거하였습니다.
#### 1차 제거 과정에서는 감탄사를 제거했고, 2차에서는 <도강록>에 해당 단어가 어떤 의미로 사용되었는지 파악 후 직접 불용어 리스트를 제작하여 제거했습니다.

```python
# 불용어 제거 과정 - 1차 기본
with open("stopwords.txt", 'r',encoding='cp949') as stopw:
    list_file = stopw.readlines()
stopwords = list_file[0].split(",")
remove_stopwords1 = [x for x in words if x not in stopwords]
c=Counter(remove_stopwords1)
c
```

```python
wc2 = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, 
                max_font_size=250,background_color='white')
gen = wc2.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)
```

```python
# 불용어 제거 과정 - 2차
# 5번 이상 사용된 명사 중 불용어 삭제
stops2 = [ '사람', '서로', '이름', '머리', '벌써', '먼저','소리', 
              '하니', '사이', '생각', '저절로', '몹시','대체', 
              '더','대개','이오','보아','스스로','바람','주부','주인',
              '부사','장관','되놈','득룡','명의','진사','일찍이',
              '크게','푸른','수십','세작','사실','잡고','갑자기',
              '위로','자리','본시','양쪽','더욱','물이','지은',
              '일행','하인','대답','말씀','지경','아주','이로',
              '마침','장군','무어','말함','애초','앞서','멀리',
              '성명','새벽','장수','선생','사정','한바탕','부어',
              '방금','가운데','대종','사서','아래','영감','세운',
              '지난해','노인','보이지','별안간','감히','홀로','대체로',
              '로서','한창','짐작','서서','서쪽','깊이','조부','마두',
              '의당','전하','조부','정필','광녕','바닥','때마침',
              '왕기','그린','어사','자못','수백','항상','날짜',
              '이내','남쪽','이르렀다','마주','하자','참으로','마부',
              '이윽고','가득','가면','이요','간다','즈음','일부러',
              '보지','이지','지고','길이']
meaningful_words = [y for y in remove_stopwords1 if not y in stops2]
c=Counter(meaningful_words)
c
```

```python
wc3 = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, 
                max_font_size=250,background_color='white')
gen = wc3.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)
```

### 박지원은 청나라로 이동하면서 벽돌, 벽돌을 굽는 가마, 기와를 중점적으로 확인했습니다.
### 무엇보다, 박지원이 두번째로 많이 사용한 단어가 <strong>'우리나라'</strong>라는 점에서 조선을 위하던 박지원의 마음을 확인할 수 있습니다.
